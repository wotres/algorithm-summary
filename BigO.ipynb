```python
# O(2^n) =>             =>  
def fibo(num): 
    if num < 2: 
        return num 
    else: 
        return fibo(num-1) + fibo(num-2)

fibo(6)
```

```python
#  O(2^n)     
def fibo(num): 
    if num < 2: 
        return num 
    else: 
        return fibo(num-1) + fibo(num-2)

print(fibo(10))

#     for  
# => O(n)   
def fiboSimple(num):
    a, b = 0, 1
    for i in range(num):
        temp = b
        b = a + b
        a = temp
    return a

print(fiboSimple(10))
```
